```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("fastfood.csv")
print(data.head())

```

```python
restaurant_counts = data["restaurant"].value_counts()

restaurant_counts.plot(kind="pie", autopct="%1.1f%%", figsize=(7,7))
plt.title("Food Items by Restaurant")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

```

```python
plt.scatter(data["total_fat"], data["calories"])
plt.xlabel("total_fat")
plt.ylabel("calories")
plt.title("Fat vs Calories Relationship")
plt.show()
```

```python
avg_cal = data.groupby("restaurant")["calories"].mean()

avg_cal.plot(kind="bar", figsize=(10,5))
plt.title("Average Calories by Restaurant")
plt.xlabel("Restaurant")
plt.ylabel("Calories")
plt.show()
```

```python
plt.hist(data["calories"], bins=20)
plt.title("Calories Distribution")
plt.xlabel("Calories")
plt.ylabel("Number of Food Items")
plt.show()
```

```python
print("Mean:", data["calories"].mean())
print("Median:", data["calories"].median())
print("Mode:", data["calories"].mode())
```

```python
data["Calorie_Level"] = pd.cut(data["calories"], bins=3, labels=["Low", "Medium", "High"])
print(data[["calories", "Calorie_Level"]].head())
```

```python
data.boxplot(column="calories")
plt.title("Boxplot of Calories")
plt.show()
```
